```python
import pandas as pd
import numpy as np
import time
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('overall-state-smoothed.csv.gz', compression='gzip')
```

```python
data
```

```python
type(data)
```

```python
data['cli'] = data['summed_n'] * data['smoothed_pct_cli'] /100
```

```python
sns.__version__
```

```python
sns.set()
```

```python
grouped = data.groupby(['date', 'state_code'])
```

```python
grouped = grouped.agg(
    summed_n=pd.NamedAgg(column="cli", aggfunc=np.sum),
)
```

```python
type(grouped)
# grouped['date'] = pd.to_datetime(grouped['date'], errors = 'coerce')
grouped.reset_index(inplace=True)
```

```python
grouped['date'] = pd.to_datetime(grouped['date'], errors = 'coerce')
grouped.set_index('date')
```

```python
sns.set(rc={'figure.figsize':(15.7,9)})
sns.lineplot(data = grouped, x = 'date', y = 'summed_n', hue = 'state_code')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```
