```python
data.sort_values('corona_result_positive', inplace=True, ascending=False)
data['corona_result_positive'].value_counts()
```

```python
dataset = data.iloc[0:1004806]
```

```python
dataset['corona_result_positive'].value_counts()
```

```python
dataset = dataset.sample(frac=1)
```

```python
dataset
```

```python
train, test = train_test_split(dataset, test_size=0.005)
```

```python
yTr = train['corona_result_positive']
yTe = test['corona_result_positive']
xTr = train[['cough', 'fever', 'sore_throat', 'shortness_of_breath', 'head_ache', 'gender_male', 'test_indication_Abroad', 'test_indication_Contact with confirmed']]
xTe = test[['cough', 'fever', 'sore_throat', 'shortness_of_breath', 'head_ache', 'gender_male', 'test_indication_Abroad', 'test_indication_Contact with confirmed']]
```

```python
model = LogisticRegression(random_state=0).fit(xTr, yTr)
```

```python
model.predict(xTe)
```

```python
model.predict_proba(xTe)
```

```python
model.score(xTe, yTe)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(xTr, yTr)
```

```python
knn_model.predict(xTe)
```

```python
knn_model.score(xTe, yTe)
```

```python
scores = np.zeros(100)
for k in range(1, 100, 1):
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(xTr, yTr)
    scores[k - 1] = knn_model.score(xTe, yTe)
```

```python
x = np.array(range(1,16,1))
sns.lineplot(x = x, y = scores[:-85])
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(min_samples_leaf = 8000)
tree.fit(xTr, yTr)
```

```python
tree.score(xTe, yTe)
```

```python
import graphviz
from sklearn.tree import export_graphviz

def draw_tree(tree):
    dot_data = export_graphviz(tree, out_file=None, 
                               feature_names=['cough', 'fever', 'sore_throat', 'shortness_of_breath', 'head_ache', 'gender_male', 'test_indication_Abroad', 'test_indication_Contact with confirmed'], 
                               class_names=['positive', 'negative'],
                               filled=True, 
                               #impurity=True,
                               rounded=True,  
                               special_characters=True,
                               proportion = True)  #trying changing proportion = False
    
    graph = graphviz.Source(dot_data)
    graph.format = 'png'
    graph.render('tree',view=True)
```

```python
draw_tree(tree)
```
